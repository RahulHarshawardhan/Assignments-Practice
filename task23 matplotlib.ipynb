```python
#Problem-1: Draw a line plot of which, the x-axis is the "Year" and the y-axis is sum of "PM2.5" of two 
#countries Iran and #China.
```

```python
df.head(3)
```

```python
iran_series = df.query('Country=="Iran"').groupby('Year')['PM2.5'].sum()
china_series = df.query('Country=="China"').groupby('Year')['PM2.5'].sum()
```

```python
iran_series
```

```python
china_series
```

```python
plt.figure(figsize=(10,10))
plt.plot(iran_series.index, iran_series.values, label='Iran', linestyle='dashed', color='red' )
plt.plot(china_series.index, china_series.values, label='China', linestyle='dashed', color='green')
plt.title("PM2.5 Over the years", fontsize=20, color = 'red')
plt.xlabel('Year')
plt.ylabel('PM2.5')
plt.xticks(df['Year'].value_counts().index)
plt.legend()
plt.grid()
plt.show()
```

```python
#Problem-2: Draw a histogram of the column "PM10" of which the y-axis represents the probability 
#(see the documentation how to draw the probability).
```

```python
df.head()
```

```python
plt.figure(figsize=(20,10))
plt.hist(df['PM10'], density = True , bins = 50, facecolor ='green')# alpha=0.6 also we can add but it will make color opaque.
plt.xlabel('Bins')
plt.ylabel('Probability')
plt.title('Histogram of PM10', fontsize=20, color ='red')
plt.grid()
plt.show()
# density is for probability
```

```python
#Problem-3: Draw a scatter plot where x-axis represents "PM2.5" and y-axis represents "PM10" for two
#countries Poland and Chile.

chile_df = df.query("Country=='Chile'")
poland_df = df.query("Country=='Poland'")
```

```python
plt.figure(figsize=(10,10))
plt.scatter(chile_df['PM2.5'],chile_df['PM10'],color='red',marker="+", label="chile")
plt.scatter(poland_df['PM2.5'], poland_df['PM10'],color='green',marker="D",label="Poland")
plt.xlabel("PM2.5")
plt.ylabel("PM10")
plt.title("PM2.5 VS PM10 for Chile And Poland", fontsize=20, color='red')
plt.legend()
plt.grid()
plt.show
```

```python
#Problem-9 Show Quarter wise Sales data for all products as multi Bar chart.
df
```

```python
df['Date']=pd.to_datetime(['2020-{}-01'.format(month) for month in df['month_number']])
```

```python
df
```

```python
final_df = df.groupby(df['Date'].dt.quarter).sum()
```

```python
final_df
```

```python
plt.figure(figsize=(10,10))
i = -1
for col in final_df.columns[1:7]:    
    plt.bar(final_df.index + i,final_df[col],width=0.15,label=col)
    i = i - 0.15

plt.xticks(final_df.index-1.4,final_df.index)
plt.xlabel('Product')
plt.ylabel('Sales')
plt.legend()
plt.show()
```

```python
#Problem-10 Plot Stacked Bar chart quarter wise for each product.
```

```python
final_df
```

```python
all_cols = []
plt.figure(figsize=(10,10))
plt.title("Stacked graph", fontsize=20, color='red')
for col in final_df.columns[1:7]:
    if len(all_cols) == 0:
        plt.bar(final_df.index,final_df[col],label=col)
    else:
        plt.bar(final_df.index,final_df[col],bottom=sum(all_cols),label=col)
    all_cols.append(final_df[col])

plt.xticks(final_df.index - 0.02, final_df.index)
plt.legend()
plt.show()
```
