# Problem-1: Make a subplots which have 2 plots.
- For the first chart, draw a scatter plot "Monitored Cap.(MW)" vs "Total Cap. Under Maintenance (MW)" of top 5 most frequent power stations. Then draw the lines which indicate the average values of these two columns. Change the colors according to the names of the Power Stations.
- For the second chart, draw a scatter plot "Monitored Cap. (MW)" vs "Actual(MU)" of the top 5 most frequent power stations. Also draw the lines which indicates the average values of these two columns. Change the colors according to the names of the Power Stations.

```python
df = pd.read_csv('PowerGeneration.csv')
```

```python
df.head()
```

```python
most_fre_ps = df['Power Station'].value_counts().head().index.tolist()
```

```python
temp_df = df[df['Power Station'].isin(most_fre_ps)]
temp_df
```

```python
temp_df
```

```python
fig,ax = plt.subplots(nrows=1,ncols=2, figsize=(15,10))
ax[0].scatter(temp_df['Monitored Cap.(MW)'], temp_df['Total Cap. Under Maintenace (MW)'], c=temp_df['Power Station'])
ax[0].axvline(temp_df['Monitored Cap.(MW)'].mean(), color='red')
ax[0].axhline(temp_df['Total Cap. Under Maintenace (MW)'].mean(), color='black')

ax[0].set_xlabel('Monitored Cap')
ax[0].set_ylabel('Top Cap under maintenance')
ax[0].set_title('Monitored Vs Under maintainence Cap', fontsize=20, color ='red')


ax[1].scatter(temp_df['Monitored Cap.(MW)'], temp_df['Actual(MU)'], c=temp_df['Power Station'])
ax[1].axvline(temp_df['Monitored Cap.(MW)'].mean(), color='red')
ax[1].axhline(temp_df['Actual(MU)'].mean(), color='black')

ax[1].set_xlabel('Monitored Cap')
ax[1].set_ylabel('Actual(MU)')
ax[1].set_title('Monitored Vs Under Actual(MU)', fontsize=20, color ='red')

```

# Problem-3: Make a 3D Surface plot of this below mathematical equation.
z=|x|−|y|

```python
x = np.linspace(-10,10,1000)
y = np.linspace(-10,10,1000)
```

```python
xx,yy= np.meshgrid(x,y)
```

```python
z = np.abs(xx)-np.abs(yy)
```

```python
fig = plt.figure(figsize=(20,10))
ax = plt.subplot(projection='3d')
p =ax.plot_surface(xx,yy,z, cmap='viridis')
fig.colorbar(p)
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

plt.show()
```

# Problem-4: Draw the 3D Contour plot of this below equation:
z=|x|−|y|

```python
plt.figure(figsize=(20,10))
ax =plt.subplot()
p =ax.contour(xx,yy,z, cmap='viridis')
plt.colorbar(p)
ax.set_xlabel('X')
ax.set_ylabel('Y')

plt.show()
```

# Problem-5: Draw a second type of Countour plot of the below equation:
z=|x|−|y|

```python
plt.figure(figsize=(20,10))
ax =plt.subplot()
p =ax.contourf(xx,yy,z, cmap='viridis')
plt.colorbar(p)
ax.set_xlabel('X')
ax.set_ylabel('Y')

plt.show()
```

# problem-8 Create a 3-D Scatter Plot using time,x,y on below synthetic data. and give color gradiant on z

```python
# Create a 3D dataset
time = np.linspace(0, 10, 100)
x = np.sin(time)
y = np.cos(time)
z = time

# Create a DataFrame from the dataset
data = pd.DataFrame({'time': time, 'x': x, 'y': y, 'z': z})
```

```python
data
```

```python
plt.figure(figsize=(20,10))
ax = plt.axes(projection='3d')

ax.scatter(data['time'], data['x'], data['y'], c=data['z'])

ax.set_xlabel('Time', color='red', fontsize=15)
ax.set_ylabel('X-Axis',color='blue', fontsize=15)
ax.set_zlabel('Y-Axis',color='green', fontsize=15)
ax.set_title('Vibration Characteristic Of a Car body', fontsize=20, color='red')
plt.show()
```

# Problem 9: Create a surface plot and the 2 types of the contour plots of the below equation.

```python
#z = sin(sqrt(x**2+y**2))
x = np.linspace(-10,10,1000)
y = np.linspace(-10,10,1000)
xx,yy = np.meshgrid(x,y)
z =np.sin(np.sqrt(xx**2+yy**2))
```

```python
fig = plt.figure(figsize=(20,10))
ax= plt.subplot(projection='3d')
ax.plot_surface(xx,yy,z, cmap='viridis')
ax.set_xlabel('X')
ax.set_xlabel('Y')
ax.set_xlabel('Z')
plt.show()
```

```python
fig = plt.figure(figsize=(15,10))
ax= plt.subplot()
ax.contour(xx,yy,z, cmap='viridis')
ax.set_xlabel('X')
ax.set_xlabel('Y')

plt.show()
```

```python
fig = plt.figure(figsize=(15,10))
ax= plt.subplot()
ax.contourf(xx,yy,z, cmap='viridis')
ax.set_xlabel('X')
ax.set_xlabel('Y')

plt.show()
```

# Problem 10: Create a surface plot and the 2 types of the contour plots of the below equation.
z=tan(log2(x2+y2)

```python
x = np.linspace(-10,10,100)
y = np.linspace(-10,10,100)
xx,yy = np.meshgrid(x,y)
z =np.tan(np.log2(xx**2+yy**2))
```

```python
fig = plt.figure(figsize=(15,10))
ax = plt.subplot(projection='3d')
ax.plot_surface(xx,yy,z, cmap='viridis')
ax.set_xlabel('X')
ax.set_xlabel('Y')
ax.set_xlabel('Z')

plt.show()
```

```python
fig = plt.figure(figsize=(15,10))
ax= plt.subplot()
ax.contour(xx,yy,z, cmap='viridis')
ax.set_xlabel('X')
ax.set_xlabel('Y')

plt.show()
```

```python
fig = plt.figure(figsize=(15,10))
ax= plt.subplot()
ax.contourf(xx,yy,z, cmap='viridis')
ax.set_xlabel('X')
ax.set_xlabel('Y')

plt.show()
```
